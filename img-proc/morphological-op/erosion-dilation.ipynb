# Erosion and dilation


```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img_sq = cv.imread('../sample-images/square.jpg')

plt.imshow(img_sq, cmap='gray')
plt.show()
```

## The easy way (using OpenCV function)
OpenCV Documentation 
* [Erosion](https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_morphological_ops/py_morphological_ops.html#erosion)
* [Dilation](https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_imgproc/py_morphological_ops/py_morphological_ops.html#dilation)

```python
kernel = np.ones((8,8),np.uint8)
erosion = cv.erode(img_sq,kernel,iterations = 1)
dilation = cv.dilate(img_sq,kernel,iterations = 1)

plt.subplot(211)
plt.imshow(erosion)
plt.subplot(212)
plt.imshow(dilation)
plt.show()
```
