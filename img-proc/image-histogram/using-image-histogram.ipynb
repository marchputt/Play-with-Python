# Image histogram
The histogram of the image can reveal important information about the image. Histogram processing (missing) can also help us achieve many useful tasks in digital image processing. 

## View the histogram
We will use `matplotlib` image histogram tool to help us visualize the histogram of different image. You can see how each histogram of each image turns out to be. 

See also: [matplotlib's histogram tutorial](https://matplotlib.org/users/image_tutorial.html#examining-a-specific-data-range)

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt


# Configure matplotlib figure size
plt.rcParams['figure.figsize'] = [15, 7]

img_natural = cv.imread('../sample-images/dog.jpg')
img_black = cv.imread('../sample-images/black.jpg')
img_white = cv.imread('../sample-images/white.jpg')
img_gray = cv.imread('../sample-images/gray.jpg')
img_grad = cv.imread('../sample-images/gradient.jpg')

img_natural_gray = cv.cvtColor(img_natural, cv.COLOR_BGR2GRAY)

# Show the images 
plt.subplot(251)
plt.imshow(img_natural_gray, cmap='gray')
plt.subplot(252)
plt.imshow(img_black, cmap='gray')
plt.subplot(253)
plt.imshow(img_white, cmap='gray')
plt.subplot(254)
plt.imshow(img_gray, cmap='gray')
plt.subplot(255)
plt.imshow(img_grad, cmap='gray')

# Now plot the histograms
plt.subplot(256)
plt.hist(img_natural_gray.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.subplot(257)
plt.hist(img_black.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.subplot(258)
plt.hist(img_white.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.subplot(259)
plt.hist(img_gray.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.subplot(2,5,10)
plt.hist(img_grad.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.show()
```

## Low and high contrast
Two images will be loaded, one of them has a nice contrast and another does not. Let's see how this reflects in the histogram differently. 

```python
castle_img = cv.imread('../sample-images/castle-gray.jpg')
castle_img_low_contrast = cv.imread('../sample-images/castle-low-contrast.jpg')

# Show the images corresponding histogram 
plt.rcParams['figure.figsize'] = [12, 10]
plt.subplot(221)
plt.imshow(castle_img, cmap='gray')
plt.subplot(222)
plt.imshow(castle_img_low_contrast, cmap='gray')
plt.subplot(223)
plt.hist(castle_img.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.subplot(224)
plt.hist(castle_img_low_contrast.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.show()
```

## Adjust the brightness

We can make the image brighter and darker by adding and subtratcing the pixel value. The brightness of the image also affects the image histogram. 

```python
img_bright = np.add(castle_img_low_contrast, 50)
img_dark = np.subtract(castle_img_low_contrast, 50)

# Show the images corresponding histogram 
plt.subplot(221)
plt.imshow(img_bright, cmap='gray')
plt.subplot(222)
plt.imshow(img_dark, cmap='gray')
plt.subplot(223)
plt.hist(img_bright.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.subplot(224)
plt.hist(img_dark.ravel(), bins=256, range=(0.0, 255), fc='k', ec='k')
plt.show()
```
